```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Read the data into Jupyter environment

```python
dia = pd.read_csv('/home/musthu/Desktop/ICT Data Science Analytics/Python practice problems/dataset/diamonds.csv')
```

```python
dia = dia.drop(dia.columns[[0]], axis=1) 
dia.head()
```

# Histogram Plot

```python
plt.hist(data=dia, x='cut', color='orange')
plt.ylabel("Counts")
plt.show()
```

# Bar Plot

```python
sns.set(rc={'figure.figsize':(13,6)})
sns.barplot(data=dia, x='cut', y='price', hue='clarity')
plt.legend()
plt.show()
```

# KDE Hist Plot

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.histplot(data=dia, x='price', color='#ee3344', kde=True)
plt.show()
```

# Joint Plot

```python
sns.jointplot(data=dia, x='price', y='carat', dropna=True, color='k')
plt.show()
```

# Correlation Matrix & Heat Map

## Correlation Matrix

```python
dia_cr = dia.drop(dia.columns[[1,2,3]], axis=1) 
dia_cr.corr()
```

## Heat Map

```python
mat = dia[['carat','depth','table','price','x','y','z']].corr()
sns.heatmap(mat, annot=True, linewidth=.3, fmt='.2f', cmap='flare',\
    annot_kws={'size': 'x-small', 'alpha': 0.75,'color':'#d3ffce'})
plt.show()
```
