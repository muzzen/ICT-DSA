# Data visualizations

## 1. Box Plot

```python

flierprops = dict(marker='d', markersize=5,\
        linestyle='none', markeredgecolor='r')

iris.boxplot(by="Classification", figsize=(8, 8),\
      color='k', flierprops= flierprops)



plt.show()
```

```python
sns.set(rc={"figure.figsize":(10, 6)}) 
sns.boxplot(x="SW", y="PW", data=iris, hue='Classification',\
    palette=['PaleGreen','YellowGreen','SeaGreen'], width=2\
    ,linewidth=1, fliersize= 4)
plt.ylabel("Petal Width")
plt.xlabel("Sepal Width")
plt.title('')

plt.show()
```

## 2. Violin Plot using Cat Plot

```python
sns.catplot(data=iris, x="SW", y="PW", \
kind='violin',hue='Classification',\
palette='turbo', legend=False, aspect=2, height=8)
plt.grid(clip_on= True, alpha= 0.3)
plt.legend()
plt.ylabel("Petal Width", fontsize=20)
plt.xlabel("Sepal Width", fontsize=20)

plt.show()
```

## 3. Count Plot

```python
sns.set(rc={"figure.figsize":(13, 6)}) 
sns.countplot(data=iris, hue='Classification', x='PL',\
    palette=['Teal','LightSeaGreen','Turquoise'])

plt.ylabel("Count", fontsize=15)
plt.xlabel("Petal Length", fontsize=15)

plt.show()
```

# 4 Joint Plot

```python
sns.jointplot(data=iris, x="PL", y="PW",\
    hue="Classification", kind="scatter", palette='cool')\
    .plot_joint(sns.kdeplot, color="r", zorder=0, levels=5)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")

plt.show()
```
